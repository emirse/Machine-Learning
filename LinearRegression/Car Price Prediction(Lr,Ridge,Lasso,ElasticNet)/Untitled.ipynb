```python
df['Model'] = df['Model'].replace({'maxda': 'mazda', 'nissan': 'Nissan', 'porcshce': 'porsche', 'toyouta': 'toyota', 
                            'vokswagen': 'volkswagen', 'vw': 'volkswagen'})
# df = df[['Model','fueltype','curbweight', 'enginesize', 'highwaympg', 'horsepower', 'citympg', 'peakrpm', 'price']]
```

```python
X = df.drop(['car_ID','price'],axis=1)
y = df['price']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
categories = list(df.select_dtypes(['object']))
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), categories)], remainder='passthrough')
X = ct.fit_transform(X)
```

```python
X
```

```python
print(y.shape)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,shuffle=True)
```

```python
alphas=[1,0.1,0.01,0.001,0.0001]
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
```

```python
def linear_reg(X_train,X_test,y_train,y_test):
    linearModel=LinearRegression()
    linearModel.fit(X_train,y_train)
    trainScore=linearModel.score(X_train,y_train).round(3)
    testScore=linearModel.score(X_test,y_test).round(3)
    print("Linear Reg Train Score:",trainScore*100,"\nLinear Reg Test Score:",testScore*100)
```

```python
linear_reg(X_train,X_test,y_train,y_test)
```

```python
def ridge_reg(X_train,X_test,y_train,y_test,alphas):
    for alpha_value in alphas:
        ridgeModel=Ridge(alpha=alpha_value,max_iter=10000).fit(X_train,y_train)
        rg_train_score=ridgeModel.score(X_train,y_train).round(3)
        rg_test_score=ridgeModel.score(X_test,y_test).round(3)
        
        print('Ridge Reg train score {} % alpha = {} '.format(rg_train_score*100,alpha_value))
        print('Ridge Reg test score {} % alpha = {} '.format(rg_test_score*100,alpha_value))
        print("----------------------------------------------------------")
    
```

```python
ridge_reg(X_train,X_test,y_train,y_test,alphas)
```

```python
def lasso_reg(X_train,X_test,y_train,y_test,alphas):
    for alpha_value in alphas:
        lassoModel=Lasso(alpha=alpha_value,max_iter=10000000).fit(X_train,y_train)
        train_score=lassoModel.score(X_train,y_train).round(3)
        test_score=(lassoModel.score(X_test,y_test)).round(3)
        
        print('Lasso Reg train score {} % alpha = {} '.format(train_score*100,alpha_value))
        print('Lasso Reg test score {} % alpha = {} '.format(test_score*100,alpha_value))
        print("----------------------------------------------------------")
    
```

```python
lasso_reg(X_train,X_test,y_train,y_test,alphas)
```

```python
def elasticnet_reg(X_train,X_test,y_train,y_test,alphas):
    
    for alpha_value in alphas:
        elasticModel=ElasticNet(alpha=alpha_value,max_iter=100000).fit(X_train,y_train)
        train_score=elasticModel.score(X_train,y_train).round(3)
        test_score=elasticModel.score(X_test,y_test).round(3)
        
        print('Elasticnet Reg train score {} % alpha = {} '.format(train_score*100,alpha_value))
        print('Elasticnet Reg test score {} % alpha = {} '.format(test_score*100,alpha_value))
        print("----------------------------------------------------------")
    
```

```python
elasticnet_reg(X_train,X_test,y_train,y_test,alphas)
```
